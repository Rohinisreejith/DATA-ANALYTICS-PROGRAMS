```python



import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import activations


print("Activation Functions")
#=================
print("Array Tensor")
x1 = tf.constant([-4.0, -2.0, 0.0, 2.0, 4.0])
print(x1)


#===========
print("1 . Linear Activation Function")
y1 = tf.keras.activations.linear(x1)
print(y1)
plt.figure(figsize=(5,5))
plt.plot(x1, y1, marker='o', mfc='r', label='Linear Activation Function')
plt.xlabel('Input Tensor')
plt.ylabel('Output Linear Activation Function')
plt.legend()
plt.grid(True)
plt.title('1. Linear Activation Function')
plt.gcf()
plt.show()

#===========
print("2 . Sigmoid Activation Function")
y2 = tf.keras.activations.sigmoid(x1)
print(y2)
plt.figure(figsize=(5,5))
plt.plot(x1, y2, marker='o', mfc='r', label='Sigmoid Activation Function')
plt.xlabel('Input Tensor')
plt.ylabel('Output Sigmoid Activation Function')
plt.legend()
plt.grid(True)
plt.title('2. Sigmoid Activation Function')
plt.gcf()
plt.show()
#===========
print("3 . Tanh Activation Function")
y3 = tf.keras.activations.tanh(x1)
print(y3)
plt.figure(figsize=(5,5))
plt.plot(x1, y3, marker='o', mfc='r', label='Tanh Activation Function')
plt.xlabel('Input Tensor')
plt.ylabel('Output Tanh Activation Function')
plt.legend()
plt.grid(True)
plt.title('3. Tanh Activation Function')
plt.gcf()
plt.show()
#===========
print("4 . Relu Activation Function")
y4 = tf.keras.activations.relu(x1)
print(y4)
plt.figure(figsize=(5,5))
plt.plot(x1, y4, marker='o', mfc='r', label='Relu Activation Function')
plt.xlabel('Input Tensor')
plt.ylabel('Output Relu Activation Function')
plt.legend()
plt.grid(True)
plt.title('4. Relu Activation Function')
plt.gcf()
plt.show()
#===========
print("5 . Leaky Relu Activation Function")
y5 = tf.keras.activations.leaky_relu(x1)
print(y5)
plt.figure(figsize=(5,5))
plt.plot(x1, y5, marker='o', mfc='r', label='Leaky Relu Activation Function')
plt.xlabel('Input Tensor')
plt.ylabel('Output Leaky Relu Activation Function')
plt.legend()
plt.grid(True)
plt.title('5. Leaky Relu Activation Function')
plt.gcf()
plt.show()
#===========
print("6 . Softmax Activation Function")
y6 = tf.keras.activations.softmax(x1)
print(y6)
plt.figure(figsize=(5,5))
plt.plot(x1, y6, marker='o', mfc='r', label='Softmax Activation Function')
plt.xlabel('Input Tensor')
plt.ylabel('Output Softmax Activation Function')
plt.legend()
plt.grid(True)
plt.title('6. Softmax Activation Function')
plt.gcf()
plt.show()

```
